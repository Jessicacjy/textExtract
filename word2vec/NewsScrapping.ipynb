```python
#paper: 新闻网站体（比如sina_paper, eastM_paper)
#i: index of news
#提取文章并且保存
def article_scrape(paper, i):

    article = paper.articles[i]
    article.download()
    article.parse()
    title = article.title
    text = article.text
    print('saving article '+title+' ...')
    file = open(title+'.txt','w')
    file.write(text)
    file.close()
```

```python
#东方财富网
eastm_paper = newspaper.build('http://www.eastmoney.com/',language = 'zh')
```

```python
#n: number of news
#paper: newspaper object
def news_scraper(n, paper_link):
    print('scraping web '+ paper_link+' ...')
    paper = newspaper.build(paper_link,language = 'zh')
    count = 0
    for i in range(n):
        try:
            article_scrape(paper, i)
            count = count + 1
        except:
            pass
```
